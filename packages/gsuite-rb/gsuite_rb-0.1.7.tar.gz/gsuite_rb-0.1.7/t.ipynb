```python
# clean the data
# df = df.dropna()
# df = df.drop_duplicates()
df = df.drop_duplicates(subset=['email'])
df = df.drop_duplicates(subset=['phone'])
df = df.drop_duplicates(subset=['name'])
df = df.drop_duplicates(subset=['email'])
```

```python
# combine head and tail of sround 5k each and save as csv
df_head = df.head(5000)
df_tail = df.tail(5000)
df_combined = pd.concat([df_head, df_tail])
df_combined.to_csv('df_combined.csv', index=False)
df_combined.shape
```
