```python
import pandas as pd
```

```python
df = pd.read_csv("eval_benchmark_4_14_25.csv")
df['query_type'].value_counts().tolist()
```

```python
import matplotlib.pyplot as plt

labels = 'Analysis', "Experiment details","Metadata structure", "Project level", "Anatomical/ Coordinates", "Subject level", "Schema documentation", "Instrument"
sizes = df['query_type'].value_counts().tolist()

fig, ax = plt.subplots()
ax.pie(sizes, labels=labels,
       colors=[
    '#FF9AA2',  # soft pink
    '#FFB7B2',  # light salmon
    '#FFDAC1',  # pale peach
    '#E2F0CB',  # light mint
    '#B5EAD7',  # seafoam green
    '#C7CEEA',  # periwinkle
    '#9FB7E0',  # soft blue
    '#D6A2E8',  # lavender
    '#F8C8DC'   # rose
])
```
