```python
import os
from pyairtable import Api
from pyairtable import api
from dotenv import load_dotenv
from fibsem_tools import read
from pyairtable.formulas import match
from cellmap_utils import upsert_image

load_dotenv()

at_api = Api(os.environ['AIRTABLE_API_KEY'])
# (switch between prod and test bases in .env)
```

```python
# get image records that are in the segmentation challenge:
image_table = at_api.table(
        os.environ["AIRTABLE_BASE_ID"], os.environ["IMAGE_TABLE_ID"]
    )
records = image_table.all(formula=match({'challenge': True, 'image_type' : 'human_segmentation'}))
```

```python
# add s3 duplicates of the segmentation challenge records

for record in records:
    ds_name = record['fields']['location'].split('/')[4]
    image_name = record['fields']['location'].split('/')[9]
    image_path = record['fields']['location']
    s3_image_path = f"s3://janelia-cosem-datasets/{ds_name}/{ds_name}.zarr/recon-1/labels/groundtruth/{image_name}/all"
    print(s3_image_path)
    image_title = 'Ground truth'
    upsert_image(at_api,
                ds_name,
                image_name,
                s3_image_path,
                image_title,
                image_type = "human_segmentation", 
                challenge=True)
```
