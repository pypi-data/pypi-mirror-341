```python
import os
import pyairtable
from pyairtable import Api
from pyairtable import api
from dotenv import load_dotenv
from fibsem_tools import read
from pyairtable.formulas import match
from cellmap_utils import upsert_image, filter_records, get_dataset_full
from pydantic import RootModel


load_dotenv()

at_api = Api(os.environ['AIRTABLE_API_KEY'])
# (switch between prod and test bases in .env)
```

```python
datasets = ['jrc_mus-dorsal-striatum-2',
'jrc_mus-cortex-3',
'jrc_mus-cerebellum-4',
'jrc_mus-cerebellum-5']
```

```python
ds_records = []
sample_records = []
acq_records = []
image_records = []
table_records = {'image' : [], 'image_acq':[], 'sample':[], 'dataset':[]}
for ds in datasets:
    img_path = f'/nrs/cellmap/data/{ds}/{ds}.zarr/recon-1/em/fibsem-int16'
    record = get_dataset_full(img_path, ds, at_api)
    table_records['dataset'].append(record['dataset'])
    table_records['sample'].append(record['sample'])
    table_records['image_acq'].append(record['image_acquisition'])
    table_records['image'].append(record['image'])
```

```python
#jsonify pydantic model records
for table_type in table_records:
    pd_type = RootModel[list[type(table_records[table_type][0])]]
    pydantic_table_records = pd_type(table_records[table_type])

    json_txt = pydantic_table_records.model_dump_json(indent=2)
    print(json_txt)
    with open(f'{table_type}_records.json', 'w') as f:
        f.write(json_txt)
    
```
